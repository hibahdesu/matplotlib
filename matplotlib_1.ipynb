```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.plot(xpoints, ypoints, marker = 'o')
plt.show()
```

```python
plt.plot(xpoints, ypoints, 'o:r')
plt.show()
```

```python
plt.plot(xpoints, ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'y')
plt.show()
```

```python
plt.plot(xpoints, ypoints, linestyle = 'dotted')
plt.show()
```

```python
plt.plot(xpoints, ypoints, linestyle = 'dashed')
plt.show()
```

```python
x = np.array([70,75, 80, 85, 90, 95, 100, 105, 110, 115])
y = np.array([240, 250, 260, 270, 280, 290, 300, 350, 400, 450])

plt.plot(x, y)

font1 = {'family': 'serif', 'color': 'red', 'size': 20}
font2 = {'family': 'serif', 'color': 'pink', 'size': 15}

plt.title('Sports Watch Data', fontdict = font1)
plt.xlabel('Average Pulse', fontdict = font2)
plt.ylabel('Calorie Burnage', fontdict = font2)

plt.show()
```
